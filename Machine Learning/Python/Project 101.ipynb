```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_path='../../../data/gasoline/'
dataFile=data_path + 'petrol_prices.html'
```

```python
from bs4 import BeautifulSoup
with open(dataFile,"rb") as f:
    soup = BeautifulSoup(f,"html.parser")
```

```python
country = []
for i in soup.findAll('a', {'class': 'graph_outside_link'}):
    i = i.text.strip().replace("*", "")
    country.append(i)
    
with open(dataFile,"rb") as f:
    soup = BeautifulSoup(f,"html.parser")
    i = soup.findAll('img', {'id' : 'commonPricesGraph'})
    price = str(i[0]).split(';')[1].split('=')[1].replace('&amp', '').split('|')
```

```python
print(len(country))
print(len(price))
```

```python
print(country[:10])
print(price[:10])
```

```python
fig = plt.figure()
fig.set_size_inches(20, 30)
plt.grid()
plt.scatter(price, country)
locs, labels = plt.xticks()
plt.setp(labels, rotation=90)
plt.tight_layout()
plt.annotate('Price in India - 1.16 $', xy=(50, 86), arrowprops=dict(facecolor='green',shrink = 0.05))
plt.savefig('1.png')
```
