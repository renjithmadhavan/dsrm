### How to find correlation ?

##### Pearson Correlation.

$$r = \frac{n\bigr(\sum xy\bigr) - \sum x \sum y}{\sqrt {[n \sum x^2 - (\sum x)^2] [n\sum y^2 - (\sum y)^2]}}$$


Example Below:



```python
x = [1, 2, 3, 4]
y = [1, 4, 9, 14]

def pearson_corr_find(x, y):
        n = len(x)
        xy = [a*b for a, b in zip(x, y)]
        sigma_xy = sum(xy)
        sigma_x = sum(x)
        sigma_y = sum(y)
        sigmax_wholesquare = sigma_x**2
        sigmay_wholesquare = sigma_y**2
        sigma_xsquare = sum([a*a for a in x])
        sigma_ysquare = sum([a*a for a in y])
        r = (n*sigma_xy - (sigma_x*sigma_y))/math.sqrt((n*sigma_xsquare - sigmax_wholesquare)*(n*sigma_ysquare - sigmay_wholesquare))
        return round(r, 4)

print(pearson_corr_find(x, y))
print(np.corrcoef(x, y)[0, 1]) ## using numpy 
print(pearsonr(x, y)[0])    ## using scipy.stats
```

```python
corr = df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

fig = plt.figure()
fig.set_size_inches(12, 10)
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask = mask, annot=True, cmap = cmap, linewidths=.5, cbar_kws={"shrink": .5}, square = True)
```
